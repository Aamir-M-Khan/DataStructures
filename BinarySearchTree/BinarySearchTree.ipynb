```python
class BinarySearchTree:
    def __init__(self, sliding_node):
        self.sliding_node = sliding_node

    def search(self, value):
        if self.sliding_node == None or self.sliding_node.value == value:
            return self.sliding_node
        elif self.sliding_node.value > value:
            return self.search(value=value)
        else:
            return self.search(value=value)

    def insert(self, value):
        if value < self.sliding_node.value:
            if self.sliding_node.left == None:
                self.sliding_node.left = TreeNode(value=value)
            else:
                self.sliding_node = self.sliding_node.left
                self.insert(value=value)

        else:
            if self.sliding_node.right == None:
                self.sliding_node.right = TreeNode(value=value)
            else:
                self.sliding_node = self.sliding_node.right
                self.insert(value=value)

    def delete(self, value):
        self.sliding_node = self._delete(self.sliding_node, value)

    def _delete(self, node, value):
        if node == None:
            return None

        if value < node.value:
            node.left = self._delete(node.left, value)
        elif value > node.value:
            node.right = self._delete(node.right, value)
        else:
            if node.left == None:
                return node.right
            elif node.right == None:
                return node.left
            node.value = self._get_min_value(node.right)
            node.right = self._delete(node.right, node.value)

        return node

    def _get_min_value(self, node):
        while node.left:
            node = node.left
        return node.value

    def succeeding_node(self, node):
        if node.left:
            new_node = node.left
            return self.succeeding_node(node=new_node)
        else:
            return node
```

```python
# Test cases
def run_tests():
    # Create a sample binary search tree
    root_node = TreeNode(50)
    bst = BinarySearchTree(root_node)

    # Insert values
    values_to_insert = [30, 70, 20, 40, 60, 80]
    for value in values_to_insert:
        bst.insert(value)

    # Search for values
    search_values = [20, 40, 60, 80, 100]
    for value in search_values:
        result = bst.search(value)
        if result:
            print(f"Value {value} found in the tree.")
        else:
            raise ValueError(f"Error: Value {value} not found in the tree.")

    # Delete values
    values_to_delete = [20, 40, 60, 80, 100]
    for value in values_to_delete:
        bst.delete(value)
        print(f"Deleted node with value {value}.")

    # Search again after deletion
    for value in search_values:
        result = bst.search(value)
        if result:
            raise ValueError(f"Error: Value {value} found in the tree after deletion.")
        else:
            print(f"Value {value} not found in the tree after deletion.")
```
